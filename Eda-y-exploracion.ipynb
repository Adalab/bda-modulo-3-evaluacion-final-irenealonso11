2.Limpieza de Datos:

```python
# DUPLICADOS:

# Calculamos el número total de filas en la columna Loyalty Number.
numero_filas = df_merge.shape[0]
print(f"Número de filas en el DataFrame: {numero_filas }")

# Calculamos el número de valores duplicados en la columna Loyalty Number.
duplicados = df_merge["Loyalty Number"].duplicated().sum()
print(f"Valores duplicados en la columna Loyalty Number: {duplicados}")

# Calculamos el porcentaje de duplicados.
porcentaje_duplicados = (duplicados / numero_filas) * 100
print(f"Porcentaje duplicado: {porcentaje_duplicados:.2f}%")
```

```python
# Visualizamos en DataFrame la frecuencia con la que se repiten los valores duplicados.
# Sacamos el número de valores duplicados en las columnas "Loyalty Number", "Year", "Month".
agrupado_duplicados = df_merge.groupby(["Loyalty Number", "Year", "Month"]).value_counts()

# Filtramos solo los valores que se repiten más de una vez.
agrupados = agrupado_duplicados[agrupado_duplicados > 1]

# Creamos un nuevo DataFrame con los valores duplicados agrupados y su frecuencia.
df_flight_duplicados = agrupados.reset_index()
df_flight_duplicados 
```

```python
# Eliminamos los duplicados filtrando por la agrupación de las 3 columnas donde se presentan esos registros duplicados.
df_merge2 = df_merge.drop_duplicates(subset=["Loyalty Number", "Year", "Month"])
df_merge2
```

```python
df_merge2.duplicated().sum()
```

```python
# Quitar el signo negativo de los valores en la columna "Salary"
df_merge2["Salary"] = df_merge2["Salary"].abs()
```

```python
#Miramos las columnas que tienen nulos
nulos_esta_num = df_merge2[df_merge2.columns[df_merge2.isnull().any()]].select_dtypes(include = np.number).columns
print("Las columnas numéricas que tienen nulos son :")
nulos_esta_num
```

```python
#Hacemos el porcentaje de nulos en las columnas que tiene nulos
df_merge2[nulos_esta_num].isnull().sum() / df_merge2.shape[0]*100
```

```python
#Sacamos la media y la mediana para analizar
media_salary = round(df_merge2['Salary'].mean(), 2)
print("media:", media_salary)
mediana_salary = df_merge2['Salary'].median()
print("mediana:", mediana_salary)
```

```python
#Hacemos la mediana a la columna de Salary ya que la mediana y la media son muy parecidas
media_salary = round(df_merge['Salary'].median(), 2)
df_merge["Salary"] = df_merge["Salary"].fillna(media_salary)

# Verificar que no queden valores nulos
nulos_after_fillna = df_merge["Salary"].isnull().sum()
print(f"Después del 'fillna', tenemos {nulos_after_fillna} valores nulos en la columna 'Salary'")
```
